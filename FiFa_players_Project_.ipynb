**Data Cleaning**

```python
#here we are converting the unmatched values from body_type to missing value first

df['body_type']=df['body_type'].str.replace('Messi','0')
df['body_type']=df['body_type'].str.replace('Courtois','0')
df['body_type']=df['body_type'].str.replace('PLAYER_BODY_TYPE_25','0')
df['body_type']=df['body_type'].str.replace('Akinfenwa','0')
df['body_type']=df['body_type'].str.replace('Shaqiri','0')
df['body_type']=df['body_type'].str.replace('Neymar','0')
df['body_type']=df['body_type'].str.replace('C. Ronaldo','0')
```

```python
#replacing '0' with missing value

df.loc[df.body_type=='0','body_type']=np.NaN
```

```python
df.isna().sum()
```

```python
#now we can see that unwanted values became missing values
dfbody=df['body_type'].value_counts()
dfbody
```

```python
#plotting again

plt.pie(dfbody,autopct='%1.1f%%',labels=dfbody.index)
```

```python
#filling the missing values of body_type column
df['body_type']=df['body_type'].fillna(df['body_type'].mode()[0])
```

```python
#to check wheather there contains any null values in our dataset
# null_values=df.isnull().sum()
# null_values

df.isna().sum()
```

```python
#encoding needed columns

from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['primary_position']=lab.fit_transform(df['primary_position'])
df['preferred_foot']=lab.fit_transform(df['preferred_foot'])
df['body_type']=lab.fit_transform(df['body_type'])
df
```

**Data Extraction**

```python
#here we can see there were many features in input,it will be easy to predict oru model result of all these feature
#so we are compaining these all features into its general features
#we are finding the weighted average of these genaral features

df['physical'] = (df["strength"] + df["sprint_speed"] + df["agility"] + df["reactions"] + df["stamina"] +
                   df["jumping"] + df["balance"] + df["acceleration"] )/8

df['defensive_skill'] = (df["sliding_tackle"] + df["standing_tackle"] + df["interceptions"] + df["marking"] +
                          df["positioning"] + df["aggression"])/6

df["football_skills"] = (df["ball_control"] + df["short_passing"] + df["long_passing"] + df["composure"] +
                          df["vision"] + df["dribbling"] + df["curve"] )/ 7

df["offensive_skills"] = (df["crossing"] + df["finishing"] + df["long_shots"] + df["volleys"] +
                           df["heading_accuracy"] + df["shot_power"] )/6

```

**Date Modeling**

```python
#creating newdataframe for needed columns only


dfnew= pd.DataFrame(data=df,columns=('age','height_cm','weight_kgs','overall_rating',
                                     'primary_position','potential',
                                     'preferred_foot','body_type','international_reputation(1-5)',
                                      'weak_foot(1-5)','skill_moves(1-5)','release_clause_euro','freekick_accuracy','penalties',
                                      'physical','defensive_skill','football_skills','offensive_skills','value_euro','wage_euro'))
dfnew
```

```python
#now we are filling the missing values in the dataset

dfnew['value_euro']=dfnew['value_euro'].fillna(dfnew['value_euro'].mean())
dfnew['wage_euro']=dfnew['wage_euro'].fillna(dfnew['wage_euro'].mean())
dfnew['release_clause_euro']=dfnew['release_clause_euro'].fillna(dfnew['release_clause_euro'].mean())
dfnew
```

```python
dfnew.isna().sum()
```

```python
#splitting x and y

x=dfnew.drop(['value_euro'],axis=1)
x
```

```python
y=dfnew['value_euro']
y
```

```python
#traning and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

**Model Development and Evaluation**

```python
#model creation and evaluation


from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error,r2_score

linear=LinearRegression()
d_tree=DecisionTreeRegressor()
r_forest=RandomForestRegressor(n_estimators=10)


lst=[linear,d_tree,r_forest]
```

```python
#we are using for loop here

for i in lst:
  print("Model is",i)
  i.fit(x_train,y_train)
  print(" "*100)

  y_pred=i.predict(x_test)
  print('Prediction Result is',y_pred)
  print(" "*100)

  mape=mean_absolute_percentage_error(y_test,y_pred)
  print('Mean absolute percentage error is',mape)
  print(" "*100)

  score=r2_score(y_test,y_pred)
  print('R2 score is',score)
  print(" "*100)

  print("*"*200)
```

```python
#creating new dataframe

newdf=pd.DataFrame({'Actual_value':y_test,'Predicted_value':y_pred,'Difference':y_test-y_pred})
newdf
```
